# Part 5: Making Decisions with if/elif/else

Programs need to make decisions. We use comparison operators and `if/elif/else`.

````python
x = 10
y = 3

print(x == y)   # equal?
print(x != y)   # not equal?
print(x > y)    # greater than?
print(x < y)    # less than?
print(x >= y)   # greater or equal?
print(x <= y)   # less or equal?
````

````python
age = 20

if age < 18:
    print("You are a child.")
elif age >= 18 and age <= 35:
    print("You are a youth.")
else:
    print("You are a senior citizen.")
````

### Logical Operators

- `and` — both conditions must be true
- `or` — at least one condition must be true
- `not` — reverses the result

### Truthiness in Python

Python treats some values as `True` and others as `False` when used in conditions.

| Value | Treated as |
|-------|-----------|
| `0`, `0.0` | `False` |
| `` (empty string) | `False` |
| `[]` (empty list) | `False` |
| `{}` (empty dictionary) | `False` |
| `None` | `False` |
| Any non-zero number | `True` |
| Any non-empty string/collection | `True` |

This is useful for checking if a variable has data before using it.

````python
# Truthiness examples
name = "Amina"
score = 0
results = []

if name:
    print("Name is provided.")

if score:
    print("Score is non-zero.")
else:
    print("Score is zero.")

if results:
    print("We have results.")
else:
    print("No results yet.")
````

### Nested Conditions

Sometimes you need to make a decision inside another decision. This is called nesting.

Use nesting when you first check one condition, then need to check another condition inside it.

````python
# Nested condition example
has_fever = True
cough_days = 5

if has_fever:
    if cough_days >= 3:
        print("Recommend malaria or pneumonia test.")
    else:
        print("Monitor at home.")
else:
    print("No fever. Likely not severe.")
````

# Part 6: Data Structures — Lists and Dictionaries

So far we have stored one value at a time. Now we will store collections of values.

### Lists

Lists store items in order. They use square brackets `[]`.

````python
fruits = ["Mango", "Pineapple", "Grapes", "Orange"]

print(fruits)
print(fruits[0])       # First item
print(len(fruits))     # Number of items

fruits.append("Avocado")
print(fruits)

fruits[1] = "Banana"
print(fruits)
````

### Dictionaries

Dictionaries store data as **key-value pairs**. They use curly braces `{}`.

Think of a real dictionary: you look up a word (key) to get the meaning (value).

````python
patient = {
    "name": "Amina Otieno",
    "age": 25,
    "bp": 120,
    "bmi": 22.5
}

print(patient)
print(patient["name"])
print(f"{patient['name']} is {patient['age']} years old.")

patient["bp"] = 118  # update value
patient["hospital"] = "Aga Khan"  # add new key
print(patient)
````

### Tuples and Sets (Briefly)

- **Tuples** are like lists but cannot change after creation. `()`
- **Sets** store only unique items. `{}` (but not key-value pairs!)

````python
coordinates = (10, 20)          # tuple
grades = {"A", "B", "A", "C"}  # set — duplicates removed

print(coordinates)
print(grades)
````

### More List Methods

Lists are very flexible. Here are the most common methods you will use.

| Method | What it does | Example |
|--------|-------------|---------|
| `.append(x)` | Add item at the end | `fruits.append("pear")` |
| `.insert(i, x)` | Insert at position `i` | `fruits.insert(1, "pear")` |
| `.remove(x)` | Remove first matching item | `fruits.remove("banana")` |
| `.pop(i)` | Remove and return item at index | `fruits.pop(0)` |
| `.sort()` | Sort in place | `fruits.sort()` |
| `.reverse()` | Reverse in place | `fruits.reverse()` |
| `.index(x)` | Find position of item | `fruits.index("mango")` |
| `.count(x)` | Count occurrences | `fruits.count("mango")` |
| `len(list)` | Number of items | `len(fruits)` |

````python
# List methods in action
fruits = ["Mango", "Pineapple", "Grapes", "Orange"]

fruits.append("Pear")
fruits.insert(1, "Banana")
fruits.sort()

print(fruits)
print(f"Number of fruits: {len(fruits)}")
print(f"Index of Grapes: {fruits.index('Grapes')}")
````

### More Dictionary Methods

Dictionaries are powerful for structured data. Learn these methods early.

| Method | What it does |
|--------|-------------|
| `.keys()` | Get all keys |
| `.values()` | Get all values |
| `.items()` | Get key-value pairs as tuples |
| `.get(key, default)` | Safely get a value (no error if missing) |
| `.update(dict)` | Add or overwrite multiple keys |
| `del dict[key]` | Remove a key |

````python
# Dictionary methods
patient = {
    "name": "Amina",
    "age": 25,
    "bp": 120
}

print("Keys:", patient.keys())
print("Values:", patient.values())
print("Items:", patient.items())

# Safe lookup
print("Diagnosis:", patient.get("diagnosis", "Not recorded"))

# Update multiple values
patient.update({"bp": 118, "bmi": 22.5})
print(patient)
````

### Which Data Structure Should I Use?

| Use this | When you need... |
|----------|------------------|
| **List** | An ordered collection that may change |
| **Tuple** | Fixed data that should not change (e.g., coordinates) |
| **Dictionary** | Labeled data with key-value pairs |
| **Set** | Unique items only, order does not matter |

If you are unsure, start with a **list** or a **dictionary**.

````python
# Example: choose the right structure

# Patient list for a waiting queue
waiting_queue = ["Amina", "Ben", "Carla"]

# Fixed lab result tuple (patient id, result, date)
lab_result = ("P001", "Negative", "2025-07-13")

# Patient dictionary with labeled info
patient = {"name": "Amina", "age": 25, "bp": 120}

# Unique set of symptoms
symptoms = {"fever", "cough", "fever", "fatigue"}

print(waiting_queue)
print(lab_result)
print(patient)
print(symptoms)
````

# Part 7: Loops

Loops let us repeat actions without writing the same code many times.

### For Loops

Use a `for` loop to go through each item in a list, string, or range of numbers.

````python
fruits = ["Mango", "Pineapple", "Grapes", "Orange"]

for fruit in fruits:
    print(f"I like {fruit}")

print("---")

for number in range(1, 6):
    print(number)
````

### While Loops

A `while` loop keeps running as long as a condition is true.

> ⚠️ Be careful — if the condition never becomes false, the loop runs forever!

````python
count = 1

while count <= 5:
    print(count)
    count += 1   # same as count = count + 1
````

### Looping Through Dictionaries

You can loop through the keys, values, or both.

````python
# Loop through a dictionary
patient = {"name": "Amina", "age": 25, "bp": 120}

print("Keys:")
for key in patient:
    print(key)

print("\nKey-value pairs:")
for key, value in patient.items():
    print(f"{key}: {value}")
````

### `enumerate()` and `zip()`

- `enumerate()` gives you both the index and the value while looping.
- `zip()` lets you loop through two lists at the same time.

````python
# enumerate example
fruits = ["Mango", "Pineapple", "Grapes"]

for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

print("---")

# zip example
names = ["Amina", "Ben", "Carla"]
ages = [25, 34, 28]

for name, age in zip(names, ages):
    print(f"{name} is {age} years old.")
````

### List Comprehensions (A Quick Preview)

A list comprehension is a shorter way to create a list from another list.

It is optional for beginners, but very common in Python.

````python
# Long way
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print(squares)

# Short way with list comprehension
squares = [n ** 2 for n in range(1, 6)]
print(squares)
````

# Part 8: Functions

A **function** is a reusable block of code. You define it once and use it many times.

## Function Basics

A **function** is a reusable block of code that performs a specific task. You define it once and call it many times.

Functions help you:
- Avoid repeating code
- Break big problems into smaller pieces
- Make your code easier to read and test

A function is created with the `def` keyword, followed by a name, parentheses `()`, and a colon `:`.

````python
def greet(name):
    """Return a greeting message."""
    return f"Hello, {name}!"

print(greet("Amina"))
print(greet("Ben"))
````

````python
def calculate_bmi(weight, height):
    """Calculate BMI from weight (kg) and height (m)."""
    return weight / (height ** 2)

my_bmi = calculate_bmi(70, 1.75)
print(f"BMI: {my_bmi:.2f}")
````

### Parameters, Arguments, and Default Values

- **Parameters** are the names in the function definition.
- **Arguments** are the actual values passed in.
- **Default values** are used when an argument is not provided.

````python
# Function with default value
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Amina"))
print(greet("Ben", "Good morning"))
````

### Local vs Global Variables

Variables inside a function are **local** — they only exist inside that function. Variables outside are **global**.

````python
# Scope example
country = "Kenya"  # global variable

def show_location():
    city = "Nairobi"  # local variable
    print(f"{city} is in {country}")

show_location()
# print(city)  # This would cause an error!
````

### Returning Multiple Values

A function can return more than one value by returning a tuple.

````python
# Multiple return values
def analyze_numbers(a, b):
    total = a + b
    product = a * b
    return total, product

sum_result, product_result = analyze_numbers(4, 5)
print(f"Sum: {sum_result}, Product: {product_result}")
````

### Docstrings

A docstring explains what a function does. It goes right after the `def` line.

````python
def calculate_bmi(weight, height):
    """
    Calculate BMI from weight (kg) and height (m).
    
    Parameters:
    weight (float): Weight in kilograms.
    height (float): Height in meters.
    
    Returns:
    float: Body Mass Index.
    """
    return weight / (height ** 2)

# You can view the docstring with help()
help(calculate_bmi)
````

## 10.2 Working with CSV Files

CSV (Comma-Separated Values) is one of the most common data formats. We can write a CSV file using plain Python and then read it with Pandas.

````python
# Create a CSV file using plain Python
import csv

patients = [
    ['name', 'age', 'bp'],
    ['Amina', 25, 120],
    ['Ben', 34, 135],
    ['Carla', 28, 122],
]

with open('patients.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(patients)

print('patients.csv created.')
````

## 10.3 Working with JSON Files

````python
import json

# Create a dictionary
patient = {
    "name": "Amina",
    "age": 25,
    "bp": 120,
    "conditions": ["none"]
}

# Write to JSON file
with open('patient.json', 'w') as file:
    json.dump(patient, file, indent=4)

print('patient.json created.')
````

# Part 11: Data Cleaning with Pandas

Real-world data is usually messy. Missing values, duplicates, and wrong column names are common problems. Pandas makes cleaning easier.

````python
# Create a messy dataset
import pandas as pd
import numpy as np

data = {
    'Patient_ID': ['P001', 'P002', 'P003', 'P004', 'P002', 'P005'],
    'Name': ['Amina', 'Ben', 'Carla', 'David', 'Ben', 'Eva'],
    'Age': [25, 34, 28, 45, 34, None],
    'BP': [120, 135, None, 140, 135, 118],
    'Diagnosis': ['Healthy', 'Malaria', 'Healthy', 'Typhoid', 'Malaria', 'Healthy']
}

df = pd.DataFrame(data)
df
````

## 11.1 Inspecting the Data

Before cleaning, always look at the data.

````python
# Shape: rows and columns
print('Shape:', df.shape)

# Info: column types and missing values
print('\nInfo:')
df.info()

# Summary statistics
print('\nStatistics:')
print(df.describe())
````

## 11.2 Handling Missing Values

Missing values can break analysis. Pandas gives us tools to find and handle them.

````python
# Find missing values
print(df.isnull().sum())
````

````python
# Drop rows with missing values
df_dropped = df.dropna()
print('After dropping missing values:')
print(df_dropped)

# Fill missing values
df_filled = df.copy()
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].median())
df_filled['BP'] = df_filled['BP'].fillna(df_filled['BP'].mean())
print('\nAfter filling missing values:')
print(df_filled)
````

## 11.3 Removing Duplicates

Duplicate rows can skew results. Use `drop_duplicates()` to remove them.

````python
# Remove duplicate rows
print('Before:', df.shape)
df_unique = df.drop_duplicates()
print('After:', df_unique.shape)
df_unique
````

## 11.4 Renaming Columns and Creating New Ones

Good column names make data easier to work with.

````python
# Rename columns
df_clean = df.copy()
df_clean = df_clean.rename(columns={'BP': 'Blood_Pressure', 'Diagnosis': 'Condition'})

# Create a new column
df_clean['Age_Group'] = df_clean['Age'].apply(lambda x: 'Adult' if x >= 30 else 'Young')

print(df_clean)
````

## 11.5 Filtering and Sorting

Filter rows based on conditions and sort by values.

````python
# Filter patients with BP above 125
high_bp = df_clean[df_clean['Blood_Pressure'] > 125]
print(high_bp)

# Sort by age
sorted_df = df_clean.sort_values('Age', ascending=False)
print('\nSorted by age:')
print(sorted_df)
````

# Part 12: Data Visualization

````python
# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(viz_df[['Age', 'BMI', 'BP']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
````

# Part 13: Numerical Computing with NumPy

NumPy is the foundation for numerical computing in Python. It provides fast, efficient arrays.

````python
import numpy as np

# Create arrays
a = np.array([1, 2, 3, 4, 5])
b = np.zeros(5)
c = np.ones((2, 3))
d = np.arange(0, 10, 2)

print('a:', a)
print('b:', b)
print('c:', c)
print('d:', d)
````

## 13.1 Array Operations

NumPy performs operations on entire arrays at once, without loops.

````python
# Operations on arrays
a = np.array([1, 2, 3, 4, 5])

print('a + 10:', a + 10)
print('a * 2:', a * 2)
print('a squared:', a ** 2)
print('a > 3:', a > 3)
````

````python
# Element-wise operations between two arrays
b = np.array([10, 20, 30, 40, 50])
print('a + b:', a + b)
print('a * b:', a * b)
````

## 13.2 Indexing and Slicing

````python
# Indexing and slicing
arr = np.array([10, 20, 30, 40, 50, 60])

print('First element:', arr[0])
print('Last element:', arr[-1])
print('Slice [1:4]:', arr[1:4])
print('Elements > 30:', arr[arr > 30])
````

## 13.3 Basic Statistics

````python
# Statistics
nums = np.array([12, 15, 18, 21, 24, 100])

print('Mean:', np.mean(nums))
print('Median:', np.median(nums))
print('Standard deviation:', np.std(nums))
print('Min:', np.min(nums))
print('Max:', np.max(nums))
````

## 13.4 NumPy vs Python Lists

NumPy arrays are much faster and use less memory than Python lists for numerical work.

````python
# Speed comparison
import time

py_list = list(range(1000000))
np_arr = np.arange(1000000)

# Python list
start = time.time()
result = [x * 2 for x in py_list]
print('Python list time:', time.time() - start)

# NumPy array
start = time.time()
result = np_arr * 2
print('NumPy array time:', time.time() - start)
````
