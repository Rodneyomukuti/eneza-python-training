# Comparison Operators in Python

**Comparison operators** compare two values and return `True` or `False`.  

| Operator | Name               | Example       | Result  | Mnemonic                |
|----------|--------------------|---------------|---------|-------------------------|
| `==`     | Equal to           | `5 == 5`      | `True`  | "Is the left **equal** to the right?" |
| `!=`     | Not equal to       | `5 != 3`      | `True`  | "Are they **different**?" |
| `>`      | Greater than       | `10 > 7`      | `True`  | "Left **larger** than right?" |
| `<`      | Less than          | `10 < 7`      | `False` | "Left **smaller** than right?" |
| `>=`     | Greater or equal   | `10 >= 10`    | `True`  | "Left **larger or equal**?" |
| `<=`     | Less or equal      | `9 <= 10`     | `True`  | "Left **smaller or equal**?" |


# Build Your Own gradeing System with Input

Write a Python program that:

1 Defines a dictionary with students’ records, where each student has:

- Name
- Score
- Cheating status (True or False)

2️ Uses input() to ask the user to type in which student’s name to gradee.

3️ Looks up that student’s record in the dictionary if absent **print record not available**.

4️ Uses if/elif/else with comparison and logical operators (and, or, not) to assign the gradee:

- A: 90–100
- B: 80–89
- C: 70–79
- D: 60–69
- F: below 60

#### BUT: if score < 60 or cheating is True ⇒ automatic "F"

5️ Prints the student's name, score, cheating status, and final letter gradee.


````python
# Define student records in a dictionary
students = {
    "Samuel": {"Score": 92, "Cheating": True},
    "Fred": {"Score": 72, "Cheating": False},
    "Jane": {"Score": 58, "Cheating": True},
    "Mary": {"Score": 95, "Cheating": False}
}

# Take student name from the user
name = input("Please enter student name: ")    # input() function

# Change to sentence case for uniformity
name = name.capitalize()        # Sting method .capitalize() # SAMUEL or samuel or sAmuEL

# Check wether the student record exists
if students.get(name, False):           # .get method for dictionaries 
    student_data = students[name]       # get the student recored -> {"Score": 72, "Cheating": False}
    score = student_data['Score']       # get the score
    status = student_data['Cheating']   # get the cheating status
    
    # gradee the score
    if score >= 90 and not status:
        grade = 'A'
    elif score >= 80 and not status:
        grade = 'B'
    elif score >= 70 and not status:
        grade = 'C'
    elif score >= 60 and not status:
        grade = 'C'
    elif score < 60 or status:
        grade = 'F'
        
    print(f"""
      Student name   : {name}
      Score          : {score}
      Cheating status: {status}
      Final grade    : {grade}
      """)
else: 
    print(f'{name} is not in our record')
````
